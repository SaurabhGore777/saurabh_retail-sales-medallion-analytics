Total_Sales

```python
%sql
SELECT
    ROUND(SUM(sales),2) AS total_sales,
    ROUND(SUM(profit),2) AS total_profit,
    SUM(quantity) AS total_quantity
FROM workspace.gold.fact_sales;
```

Sales by Category

```python
%sql
SELECT
    dp.category,
    ROUND(SUM(fs.sales),2) AS total_sales,
    ROUND(SUM(fs.profit),2) AS total_profit
FROM workspace.gold.fact_sales fs
JOIN workspace.gold.dim_product dp
ON fs.product_key = dp.product_key
GROUP BY dp.category
ORDER BY total_sales DESC;
```

Top 10 Products

```python
%sql
SELECT
    dp.product_name,
    ROUND(SUM(fs.sales),2) AS total_sales
FROM workspace.gold.fact_sales fs
JOIN workspace.gold.dim_product dp
ON fs.product_key = dp.product_key
GROUP BY dp.product_name
ORDER BY total_sales DESC
LIMIT 10;
```

top 10 customers

```python
%sql
SELECT
    dc.customer_name,
    ROUND(SUM(fs.sales),2) AS total_sales,
    ROUND(SUM(fs.profit),2) AS total_profit
FROM workspace.gold.fact_sales fs
JOIN workspace.gold.dim_customer dc
ON fs.customer_key = dc.customer_key
GROUP BY dc.customer_name
ORDER BY total_sales DESC
LIMIT 10;
```
